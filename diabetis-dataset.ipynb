```python
#importing dependancies
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn import svm
```

```python
#loading dataset into pandas dataframe
dataset=pd.read_csv('/kaggle/input/pima-indians-diabetes-database/diabetes.csv')
```

```python
#printing first five rows
dataset.head()
```

```python
#printing last 5 rows
dataset.tail()
```

```python
#checking no.of rows and columns
dataset.shape
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
#checking for null values
dataset.isnull().sum()
```

```python
x=dataset.drop(columns='Outcome',axis=1)
y=dataset.Outcome
```

```python
#data standardization
scaler=StandardScaler()
```

```python
f_scaler=scaler.fit_transform(x)
```

```python
print(f_scaler)
```

```python
x=f_scaler
y=dataset.Outcome
```

```python
x.shape
```

```python
print(x)
```

```python
print(y)
```

```python
#train test split the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2,stratify=y)
```

```python
x.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#training the machine
reg=LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
#model evaluation
accuracy=reg.predict(x_train)
x_train_eval=accuracy_score(accuracy,y_train)
print("Accuracy score of x_train is:",x_train_eval)
```

```python
aaccuracy=reg.predict(x_test)
x_test_eval=accuracy_score(aaccuracy,y_test)
print("Accuracy score of x_test is:",x_test_eval)
```

```python
#making a predictive system
input_data=(11,143,94,33,146,36.6,0.254,51)
```

```python
arr=np.asarray(input_data)
```

```python
reshapee=arr.reshape(1,-1)
```

```python
std_data=scaler.transform(reshapee)
print(std_data)
```

```python
prediction=reg.predict(std_data)
print(prediction)
if predic[0]==0:
    print('person has no diabetis')
else:
    print('prtson has diabetis')
```
